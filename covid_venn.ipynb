```python
import pandas as pd
import upsetplot
```

```python
df = pd.read_csv('covid_venn.csv').fillna(0)
df
```

```python
grouped = df.groupby(['Anosmia', 'Cough', 'Fatigue', 'Diarrhoea', 'Breath shortness', 'Fever']).agg({'count': 'sum'})
```

```python
upsetplot.plot(grouped['count'], sort_by='cardinality')
```
